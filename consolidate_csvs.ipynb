```python
import os
import pandas as pd
```

```python
protein_files = []
for root, dirs, files in os.walk("dataset_2"):
    for file in files:
        if file.endswith(".csv"):
             protein_files.append(os.path.join(root, file).replace('\\', '/'))
                
```

```python
protein_files
```

```python
df = pd.concat((pd.read_csv(f) for f in protein_files))
```

```python
protein_list = df['name'].tolist()
```

```python
protein_list
```

```python
len(protein_list)
```

```python
# Filter uniques
protein_list_uniq = list(set(protein_list))
```

```python
protein_list_dali = []
for p in protein_list_uniq:
    protein_list_dali.append(p[0:4].lower() + p[4])
```

```python
len(protein_list_dali)
```

```python
# Write the file
f = open('dali_query_list.txt', 'w')
f.writelines([p + '\n' for p in protein_list_dali])
```
